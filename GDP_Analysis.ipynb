```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the dataframe from the file
df=pd.read_csv("Data1A.csv")
df.head(10)
```

```python
#Renaming the columns
df.rename(columns={"Items  Description":"Items_Description",'West Bengal1':'West_Bengal','All_India GDP':'All_India_GDP'},inplace=True)
df.columns
df
```

```python
#checking the errors in data frame
print(df.isnull().sum(axis=0))
print(df.isnull().values.any())
print(df.isnull().values.sum())
print(len(df))
```

```python
#droppingt the westbengal as it has only nan values 
df=df.dropna(axis=1,how='all')
print(df)
```

```python
# drooppin the columnn of duaration of 2016-17 as it is not required
index = df[(df['Items_Description'] == 'GSDP - CURRENT PRICES (` in Crore)') & ( df['Duration'] == '2016-17' )].index
df.drop(index,inplace=True)
```

```python
# drooppin the columnn of duaration of 2016-17 as it is not required
index=df[(df['Items_Description'] == '(% Growth over previous year)') & ( df['Duration'] == '2016-17' )].index
df.drop(index,inplace=True)
```

```python
#Transform the datafram
new_header = df['Items_Description']+' '+df['Duration']
new_df=df.T
new_df=new_df[2:]
new_df.columns=new_header
print(new_df.isna().sum(axis=1))
```

```python
new_df['avg_growth']=new_df[['(% Growth over previous year) 2012-13','(% Growth over previous year) 2013-14','(% Growth over previous year) 2014-15','(% Growth over previous year) 2015-16']].mean(axis=1)
new_df
```

```python
#calculating the average growth rate
avg_nat_growth=new_df.iloc[-1]['avg_growth']
new_df=new_df.drop(['All_India_GDP'])
```

```python
new_df=new_df.sort_values(by='avg_growth')
sns.barplot(x='avg_growth',y=new_df.index,data=new_df)
```

```python
#Curiosity the native of mine is bihar
avg_growth_Bihar=new_df.loc['Bihar']['avg_growth']
print(avg_growth_Bihar)
print(avg_growth_Bihar/avg_nat_growth)
```
